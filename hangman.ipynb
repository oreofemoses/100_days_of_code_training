```python
from IPython.display import clear_output
import random
hangman_stages = {
    0: """
    +---+
        |
        |
        |
        |
        |
   =======""",
    1: """
    +---+
    |   |
        |
        |
        |
        |
   =======""",
    2: """
    +---+
    |   |
    O   |
        |
        |
        |
   =======""",
    3: """
    +---+
    |   |
    O   |
    |   |
        |
        |
   =======""",
    4: """
    +---+
    |   |
    O   |
   /|   |
        |
        |
   =======""",
    5: """
    +---+
    |   |
    O   |
   /|\  |
   /    |
        |
   =======""",
    6: """
    +---+
    |   |
    O   |
   /|\  |
   / \  |
        |
   ======="""
}
random_words = [
    "Apple", "Banana", "Cat", "Dog", "Elephant", "Fish", "Giraffe", "House", "Ice", "Jaguar",
    "Kite", "Lemon", "Mountain", "Notebook", "Orange", "Pizza", "Quail", "Rainbow", "Snake", "Tree",
    "Umbrella", "Violin", "Watermelon", "Xylophone", "Yacht", "Zebra", "Apricot", "Bicycle", "Camera", "Dinosaur",
    "Eagle", "Frog", "Gorilla", "Hat", "Igloo", "Jellyfish", "Kangaroo", "Lamp", "Monkey", "Nest",
    "Octopus", "Parrot", "Quilt", "Robot", "Sun", "Turtle", "Unicorn", "Volcano", "Whale", "Xenon",
    "Yard", "Zipper", "Almond", "Bridge", "Candle", "Desert", "Engine", "Forest", "Guitar", "Helicopter",
    "Island", "Jewel", "Kettle", "Ladder", "Meadow", "Necklace", "Ocean", "Piano", "Quiver", "River",
    "Star", "Train", "Universe", "Valley", "Windmill", "X-ray", "Yawn", "Zebra", "Anchor", "Balloon",
    "Castle", "Dragon", "Envelope", "Feather", "Garden", "Horizon", "Iceberg", "Jungle", "Knight", "Lighthouse",
    "Mountain", "Nucleus", "Oasis", "Penguin", "Quest", "Rocket", "Stadium", "Thunder", "Vulture", "Wizard"
]


word =random.choice(random_words).lower() #Get a random word from our list and convert the word to lowercase.
fails = 0
display = ["_ " for i in range(len(word))] #a list of underscores each representing a letter to be guessed
guessed_letters = ''

while fails < 6 and '_ ' in display:
    clear_output(wait=True)
    
    letter = input("Guess a letter: ")
    
    if letter in guessed_letters:
        print("You've already guessed that one!")
        
    elif letter.lower() in word:
        for i in range(len(word)):
            if letter == word[i]: display[i] = letter + ' '
        print("Correct Guess!!")

    else:
        print("Wrong!!")
        fails += 1 
        
    guessed_letters += letter
    print(f"{''.join(display)}{hangman_stages[fails]}")
    
    
if '_ ' not in display:#ie no unguessed letter at the end of the run
    print(f"YOU WIN!!! The correct word is {word}")
else: print(f"YOU LOOSE!, WE'LL GET THEM NEXT TIME... The correct word is {word}")
```
