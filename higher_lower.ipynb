```python
import random
from IPython.display import clear_output

correct = True
randomized_index = random.sample(range(0, len(data)), len(data)) # we need to extract data randomly to add a bit of complexity
index = 0

while correct and index < len(randomized_index)-1:
    celeb_A = data[randomized_index[index]] #get a random index from randomized index in order.
    celeb_B = data[randomized_index[index + 1]]
    
    print(logo)
    print(f"Compare A: {celeb_A['name']}, a {celeb_A['description']}, from {celeb_A['country']}: {celeb_A['follower_count']}")
    print(vs)
    print(f"Compare B: {celeb_B['name']}, a {celeb_B['description']}, from {celeb_B['country']}: {celeb_B['follower_count']}")
    
    answer = input("Who has more followers? Type 'A' or 'B': ").upper()
    
    answer = {'A': celeb_A['follower_count'], 'B': celeb_B['follower_count']}.get(answer, 0)
    
    if answer == max(celeb_A['follower_count'], celeb_B['follower_count']):
        print("Correct!")
        index +=1
    else:
        print('Wrong, You loose')
        correct = False
    clear_output(wait=True)
    
print(f"Score {index}")
```
